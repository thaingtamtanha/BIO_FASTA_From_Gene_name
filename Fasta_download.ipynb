```python
x = ['ACTB', 'AUTS2', 'CKB', 'FKBP8']
animal = 'Homo sapien'
out_handle = open('Result_fasta.fasta', 'w')

for item in x:
    print(str(item))
    search_string = str(item)+"[Gene] AND Homo sapien [Organism] AND mRNA[Filter] AND RefSeq[Filter]"
    handle = Entrez.esearch(db="nucleotide", term=search_string)
    record = Entrez.read(handle)
    ids = record['IdList']
    seq_id = ids[0]
    handle = Entrez.efetch(db="nucleotide", id=seq_id, rettype="fasta", retmode="text")
    record = handle.read()
    test = str(record)
    #print(record)
    new_string = test.replace("T", "U")
    print(new_string)
    out_handle.write(new_string.rstrip('\n'))
```
